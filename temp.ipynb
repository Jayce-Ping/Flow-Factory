```python
import math
batch_size = 4
group_size = 10                # Number of repetitions per sample
unique_sample_num = 19                    # `Least` number of unique sample per epoch
num_replicas = 6      # Total number of replicas, process num, gpu num
rank = 0              # Current replica rank
seed = 42              # Random seed for synchronization

# Compute the number of samples for each batch iteration
sample_num_per_iteration = num_replicas * batch_size
step = sample_num_per_iteration // math.gcd(group_size, sample_num_per_iteration)
new_m = (unique_sample_num + step - 1) // step * step  # Round up m to be multiple of step
if new_m != unique_sample_num:
    print(f"Adjusted m from {unique_sample_num} to {new_m} to make sure `unique_sample_num`*`group_size` is multiple of `batch_size`*`num_replicas` for even distribution.")
    unique_sample_num = new_m

num_batches_per_epoch = (unique_sample_num * group_size) // sample_num_per_iteration
num_batches_per_epoch
```
